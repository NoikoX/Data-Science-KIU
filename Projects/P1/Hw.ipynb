```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# In this project i will use NumPy functions to manipulate an image, the tasks include rotating,
# cropping, adjusting brightness, and applying mathematical operations on the image using NumPy
```

```python
def display_image(image, title="Image"):
    plt.imshow(image)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
image_path = "Images/Pawawo.jpg"
image = Image.open(image_path)
image_np = np.array(image)
```

```python
display_image(image_np, "Original Image")
```

```python
rotated_90 = np.rot90(image_np, 1)
display_image(rotated_90, "Rotated 90 Degrees")
```

```python
rotated_180 = np.rot90(image_np, k=2)
display_image(rotated_180, "Rotated 180 degrees")
```

```python
rotated_270 = np.rot90(image_np, k=3)
display_image(rotated_270, "Rotated 270 degrees")
```

```python
Image.fromarray(rotated_90).save("Images/rotated_90.jpg")
Image.fromarray(rotated_180).save("Images/rotated_180.jpg")
Image.fromarray(rotated_270).save("Images/rotated_270.jpg")
```

```python
def adjust_brightness(image_np, f):
    bright_img = np.clip(image_np * f, 0, 255).astype(np.uint8)
    return bright_img
```

```python
brighter = adjust_brightness(image_np, 1.5)
display_image(brighter, "Brighter Image 50%")
```

```python
darker = adjust_brightness(image_np, 0.5)
display_image(darker, "Darker Image 50%")
```

```python
Image.fromarray(brighter).save("Images/BrighterImage.jpg")
Image.fromarray(darker).save("Images/DarkerImage.jpg")
```

```python
def crop_center(image_np, crop_fraction = 0.5):
    h, w, _ = image_np.shape
    new_h, new_w = int(h * crop_fraction), int(w * crop_fraction)
    start_h = (h - new_h) // 2
    start_w = (w - new_w) // 2
    cropped_image = image_np[start_h : start_h + new_h, start_w : start_w + new_w]
    return cropped_image
```

```python
cropped_image = crop_center(image_np, crop_fraction=0.5)
display_image(cropped_image, "Cropped central image with 50%")
```

```python
cropped_image2 = crop_center(image_np, crop_fraction=0.2)
display_image(cropped_image2, "Cropped central image with 20%")
```

```python
Image.fromarray(cropped_image).save("Images/CroppedImage.jpg")
```

```python
mean_value = np.mean(image_np)
max_value = np.max(image_np)
min_value = np.min(image_np)
sum_value = np.sum(image_np)

print(f"Mean pixel value: {mean_value}")
print(f"Max pixel value: {max_value}")
print(f"Min pixel value: {min_value}")
print(f"Sum of all pixel values: {sum_value}")
```

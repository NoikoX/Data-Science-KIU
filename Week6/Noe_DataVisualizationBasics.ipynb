```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('company_sales_data.csv')

df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit per Month')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['total_units'], linestyle=':', color='red', marker='o', markerfacecolor='red', linewidth=3, label='Total Units Sold')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='lower right')
plt.title('Total Units Sold per Month')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['facecream'], label='Face Cream')
plt.plot(df['month_number'], df['facewash'], label='Face Wash')
plt.plot(df['month_number'], df['toothpaste'], label='Toothpaste')
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap')
plt.plot(df['month_number'], df['shampoo'], label='Shampoo')
plt.plot(df['month_number'], df['moisturizer'], label='Moisturizer')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Product Sales per Month')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['month_number'], df['toothpaste'], label='Toothpaste Sales')
plt.grid(True, linestyle='--')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales per Month')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
bar_width = 0.35
plt.bar(df['month_number'] - bar_width/2, df['facecream'], width=bar_width, label='Face Cream', align='center')
plt.bar(df['month_number'] + bar_width/2, df['facewash'], width=bar_width, label='Face Wash', align='center')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Face Cream and Face Wash Sales per Month')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['month_number'], df['bathingsoap'], label='Bathing Soap')
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.title('Bathing Soap Sales per Month')
plt.savefig('bathingsoap_sales.png')  # Save the image (because this plot is too pretty not to!)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['total_profit'], bins=5)
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.title('Profit Distribution')
plt.show()
```

```python
total_sales = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()]
products = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
plt.figure(figsize=(10, 5))
plt.pie(total_sales, labels=products, autopct='%1.1f%%')
plt.title('Total Sales per Product for Last Year')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap Sales')
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.title('Bathing Soap Sales per Month')
plt.subplot(2, 1, 2)
plt.plot(df['month_number'], df['facewash'], label='Face Wash Sales', color='orange')
plt.xlabel('Month Number')
plt.ylabel('Face Wash Sales')
plt.title('Face Wash Sales per Month')
plt.tight_layout()
plt.show()
```

```python
#somethingg like that
plt.figure(figsize=(10, 5))
plt.stackplot(df['month_number'], df['facecream'], df['facewash'], df['toothpaste'], df['bathingsoap'], df['shampoo'], df['moisturizer'],
              labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('All Product Sales')
plt.legend()
plt.show()
```
